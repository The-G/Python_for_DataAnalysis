- 'https?' => 'http' 'https'
- '\S' => space 제외 모든문자.
- 대문자가 주로 반대이네,

```python
chunker = pd.read_csv('ch06/ex6.csv', chunksize=1000) # chunker를 이용해서 1000개씩!

tot = Series([])
for piece in chunker: # 10개 덩어리에서 1개씩 넘어간다. 그 1 덩어리가 1000개!
    tot = tot.add(piece['key'].value_counts(), fill_value=0) # 값이 없으면 0
                                                             # key값에 대한 wordCouning을 한거네!!!

tot = tot.sort_values(ascending=False) # 내림차순 정렬
```

- piece는 dataframe의 한 줄 넘어오게 된다.

이렇게 데이터가 선택되지 않고 사라지는 경우가 있다

```python
class my_dialect(csv.Dialect): # 사용자 정의로 하네!!
    lineterminator = '\n' # 줄종료
    delimiter = ';'       # 줄구분
    quotechar = '"'       # 인용부호
    quoting = csv.QUOTE_MINIMAL
```

```python
with open('mydata.csv', 'w') as f:
    writer = csv.writer(f, dialect=my_dialect)
    writer.writerow(('one', 'two', 'three'))
    writer.writerow(('1', '2', '3'))
    writer.writerow(('4', '5', '6'))
    writer.writerow(('7', '8', '9'))
```

```python
from lxml.html import parse
from urllib2 import urlopen

parsed = parse(urlopen('http://finance.yahoo.com/q/op?s=AAPL+Options'))

doc = parsed.getroot()
```

```python
links = doc.findall('.//a')
links[15:20]
```

```python
lnk = links[28]
lnk
lnk.get('href')
lnk.text_content()
```

```python
urls = [lnk.get('href') for lnk in doc.findall('.//a')]
urls[-10:]
```

```python
tables = doc.findall('.//table')
calls = tables[9]
puts = tables[13]
```

```python
rows = calls.findall('.//tr')
```

```python
def _unpack(row, kind='td'):
    elts = row.findall('.//%s' % kind)
    return [val.text_content() for val in elts]
```

```python
_unpack(rows[0], kind='th')
_unpack(rows[1], kind='td')
```

```python
from pandas.io.parsers import TextParser

def parse_options_data(table):
    rows = table.findall('.//tr')
    header = _unpack(rows[0], kind='th')
    data = [_unpack(r) for r in rows[1:]]
    return TextParser(data, names=header).get_chunk()
```

```python
call_data = parse_options_data(calls)
put_data = parse_options_data(puts)
call_data[:10]
```

#### Parsing XML with lxml.objectify

```python
!type ch06\mta_perf\Performance_MNR.xml
```

```python
from lxml import objectify # xml을 읽어서 객체화 시키겠다!!

path = 'ch06\mta_perf\Performance_MNR.xml'
parsed = objectify.parse(open(path))
parsed
root = parsed.getroot()
root
```

- xml 할 때 DOM이라는 개념을 쓴다!

```python
data = []

skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.INDICATOR: # root의 INDICATOR 부터 시작 하겠다!!
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval # pyval이 뭐지?!
    data.append(el_data)
```

## Interacting with databases, sqllite와 연결

```python
import sqlite3

query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER
);"""

con = sqlite3.connect(':memory:')
con.execute(query)
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
        ('Tallahassee', 'Florida', 2.6, 3),
        ('Sacramento', 'California', 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"

con.executemany(stmt, data) # insert 3q번 수행되지!!
con.commit()
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall() # 읽어 오겠다
rows
```
